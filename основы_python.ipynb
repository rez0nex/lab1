#Задача 1
Написать функцию на вход которой подается строка, состоящая из латинских букв.
Функция должна вернуть количество гласных букв (a, e, i, o, u) в этой строке. Написать тесты для кода

```python
def count_vowels(s):
    return len([char for char in s.lower() if char in 'aeiou'])

print(count_vowels('hello'))    # 2
print(count_vowels('HELLO'))    # 2
print(count_vowels('abc'))      # 1
print(count_vowels('aeiou'))    # 5
print(count_vowels('xyz'))      # 0
```

# Задача 2
Написать функцию на вход, которой подается строка. Функция должна вернуть true, если
каждый символ в строке встречается только 1 раз, иначе должна вернуть false. Написать тесты для кода

```python
def is_unique(s):
    return len(s) == len(set(s))

print(is_unique('hello'))   #False
print(is_unique('HELLO'))   #False
print(is_unique('abc'))     #True 
print(is_unique('aabbcc'))  #False
print(is_unique(''))        #True
```

# Задача 3
Написать функцию, которая принимает положительное число и возвращает количество
бит равных 1 в этом числе. Написать тесты для кода

```python
def count_ones(n):
    return bin(n)[2:].count('1')

print(count_ones(7))    # 3  (7 = 111)
print(count_ones(8))    # 1  (8 = 1000)
print(count_ones(15))   # 4  (15 = 1111)
print(count_ones(0))    # 0  (0 = 0)
print(count_ones(255))  # 8  (255 = 11111111)
```

# Задача 4
Написать функцию, которая принимает положительное число. Функция должна вернуть то,
сколько раз необходимо перемножать цифры числа или результат перемножения, чтобы
получилось число состоящее из одной цифры.
Например, для входного числа:
· 39 функция должна вернуть 3, так как 3*9=27 => 2*7=14 => 1*4=4
· 4 функция должна вернуть 0, так как число уже состоит из одной цифры
· 999 функция должна вернуть 4, так как 9*9*9=729 => 7*2*9=126 => 1*2*6=12 =>
1*2=2. Написать тесты для кода

```python
def magic(n):
    steps = 0
    while n >= 10:
        res = 1
        while n > 0:
            res *= n % 10
            n //= 10
        n = res
        steps += 1
    return steps

print(magic(39))    # 3  (39 => 27 => 14 => 4, 3 шага)
print(magic(999))   # 4  (999 => 729 => 126 => 12 => 2, 4 шага)
print(magic(4))     # 0  (4, 0 шагов)
print(magic(25))    # 2  (25 => 10 => 0, 2 шага)
print(magic(123))   # 1  (123 => 6, 1 шаг)
```

# Задача 5
Написать функцию, которая принимает два целочисленных вектора одинаковой длины и
возвращает среднеквадратическое отклонение двух векторов. Написать тесты для кода

```python
def mse(pred, true):
    return sum([(pred[i] - true[i]) ** 2 for i in range(len(pred))]) / len(pred)

print(mse([1, 2, 3], [1, 2, 3]))           # 0.0
print(mse([1, 2, 3], [2, 3, 4]))           # 1.0
print(mse([0, 0, 0], [1, 1, 1]))           # 1.0
print(mse([1, 2, 3, 4], [2, 3, 4, 5]))     # 1.0
print(mse([10, 20, 30], [15, 25, 35]))     # 25.0
```

# Задача 6
Написать функцию, принимающая целое положительное число. Функция должна вернуть
строку вида “(n1**p1)(n2**p2)…(nk**pk)” представляющая разложение числа на простые
множители (если pi == 1, то выводить только ni).
Например, для числа 86240 функция должна вернуть “(2**5)(5)(7**2)(11)”. Написать тесты для кода

```python
def factorize(n):
    if n == 1:
        return '(1)'
    
    factors = []
    div = 2
    
    while div * div <= n:
        count = 0
        while n % div == 0:
            count += 1
            n //= div
        if count > 0:
            if count == 1:
                factors.append(f'({div})')
            else:
                factors.append(f'({div}**{count})')
        div += 1
    
    if n > 1:
        factors.append(f'({n})')
    
    return ''.join(factors)

print(factorize(86240))   # "(2**5)(5)(7**2)(11)"
print(factorize(2025))    # "(3**4)(5**2)"
print(factorize(17))      # "(17)"
print(factorize(100))     # "(2**2)(5**2)"
print(factorize(1))       # "(1)"
```

# Задача 7
Написать функцию, принимающая целое число n, задающее количество кубиков. Функция
должна определить, можно ли из данного кол-ва кубиков построить пирамиду, то есть
можно ли представить число n как 1^2+2^2+3^2+…+k^2. Если можно, то функция должна
вернуть k, иначе строку “It is impossible”. Написать тесты для кода

```python
def pyramid(number):
    k = 1
    total = 0
    
    while total < number:
        total += k ** 2
        if total == number:
            return k
        k += 1
    
    return 'It is impossible'

print(pyramid(1))     # 1  (1)
print(pyramid(5))     # 2  (1 + 4 = 5)
print(pyramid(14))    # 3  (1 + 4 + 9 = 14)
print(pyramid(30))    # 4  (1 + 4 + 9 + 16 = 30)
print(pyramid(55))    # 5  (1 + 4 + 9 + 16 + 25 = 55)
print(pyramid(10))    # It is impossible
print(pyramid(100))   # It is impossible
```

# Задача 8
Функция принимает на вход положительное число и определяет является ли оно сбалансированным, т.е. сумма цифр до средних равна сумме цифр после. Средними в случае нечетного числа цифр считать одну цифру, в случае четного - две средних. Написать тесты для кода

```python
def is_balanced(n):
    num = str(n)
    length = len(num)
    
    if length % 2 == 0:
        mid = length // 2
        left_sum = sum(int(d) for d in num[:mid])
        right_sum = sum(int(d) for d in num[mid:])
    else:
        mid = length // 2
        left_sum = sum(int(d) for d in num[:mid])
        right_sum = sum(int(d) for d in num[mid+1:])
    
    return left_sum == right_sum

print(is_balanced(1234))      # False
print(is_balanced(123321))    # True
print(is_balanced(12321))     # True
print(is_balanced(1))         # True
print(is_balanced(123))       # False
print(is_balanced(121))       # True
```
